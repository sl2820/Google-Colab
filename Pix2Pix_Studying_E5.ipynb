#### Iterating Trained Images

```
import os.path
from IPython.display import Image
from IPython.display import display

Epoch_size = 5
Batch_size = 400

for i in range (Epoch_size):
  step = 4 # Every Five Epoch
  if i % step ==0:
    for j in range (Batch_size):
      step_j = 199 # 2 Results per Epoch
      if (j%step_j==0):
        fname = "images/facades/"+str(i)+"_"+str(j)+".png"
        if os.path.isfile(fname): # Check to see if the file exists --> prevent error
          print ("THIS IS RESULT OF Epoch: %d and Batch: %d" %(i+1,j))
          display(Image(filename="images/facades/%d_%d.png" % (i,j)))
    print("-----------------------End of Epoch: %d----------------------\n\n" % (i+1))
  
```
